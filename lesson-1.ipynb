Тип БД 
(РБД) Реалиционная база данных - структора данных, основана на табличной модели где данные хранятся в виде таблиц средами и столбцами. Каждый ряд в таблице имеет уникальный идентификатор. 

Схема данных РБД- Имеет жесткую схему данных. Структура таблицы и типы данных предопределены.

Масштабируемость
Горизонтальное масштабирование проблематично т.к Работает на одном сервере.

Сложность запросов
Использует язык структурированных запросов.

Обеспечивает жесткую поддержку транзакций. 

(NoSQL) Нереалиционная база данных - база данных может хранить данные в различных форматах : ключ-значения, графы, документы. 

Схема данных является гибкой, изменяется по мере необходимости, без ограничений.
Использует специальные API и языки запросов

Масштабируемость
Разработаны для горизонтального масштабирования, для добавления новых серверов.

Ограниченая поддержка транзакций или вовсе не поддерживает.

Символьные данные - могут хранится как строка (фиксированной) или динамической длинны.
CHAR(20) - строка фиксированной длины (выделяется память под 20 символов)(максимум 255 символов)
VARCHAR(20) - строка переменной длины (память выделяется динамически, но не более 20 символов)(0 - 65 535)
